morphing


```python
import numpy as np
```

```python
imagem1 = cv2.imread("imagem1.jfif")
imagem2 = cv2.imread("imagem2.jfif")
```

```python
if imagem1.shape != imagem2.shape:
  altura, largura, canais = imagem1.shape
  imagem2 = cv2.resize(imagem2, (largura, altura))
```

```python
alpha_values = np.linspace(0,1,10)

print(alpha_values)
```

```python
for alpha in alpha_values:
  img_morphed = cv2.addWeighted(imagem1, alpha, imagem2, 1 - alpha, 0)
  cv2_imshow(img_morphed)
```
